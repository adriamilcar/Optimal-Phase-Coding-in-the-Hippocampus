```python
from brian2 import *
prefs.codegen.target = 'numpy'

import itertools
import numpy as np
import scipy.stats
import matplotlib.cm as cm
from matplotlib.lines import Line2D
import seaborn as sns

from encoder_class.phase_encoder import PhaseEncoder, run_simulation, get_distr_params
from encoder_class.theoretical_functions import phi_of_I_lin, phi_of_I, approx_mutual_I, bins_mutual_I
from encoder_class.encoder_functions import model_sigma, min_phi_distribution
from encoder_class.plot_functions import plt_fig_3b, plt_fig_3c, plt_fig_3d, plt_fig_3e
```

```python
M = 10
N = 50

model_params = {}
model_params["tau_m"] = 24*ms
model_params["R_m"] = 142e6*ohm
model_params["v_thres"] = 15*mV
model_params["v_rest"] = 0*mV
model_params["v_reset"] = 0*mV
model_params["tau_ref"] = 0*ms
model_params["v_0"] = 0*mV
model_params["noise_frac"] = 0.2


oscillation_params = {}
oscillation_params["I_osc"] = 40*pA
oscillation_params["f"] = 1*Hz

input_params = {}
input_params["I_min"] = 75*pA
input_params["I_max"] = 130*pA
input_params["automatic_range"] = True

simulation_params = {}
simulation_params["method"] = "euler"
simulation_params["num_oscillations"] = 40
simulation_params["monitor_spikes"] = True
simulation_params["monitor_voltage"] = False
simulation_params["dt"] = 0.05*ms
simulation_params["record_dt"] = 0.5*ms


encoder = PhaseEncoder(num_ensembles=M, ensemble_size=N, model_params=model_params, 
                       oscillation_params=oscillation_params, input_params=input_params,
                       simulation_params=simulation_params, rnd_seed=0)

exp_phis = run_simulation(encoder, mode='experimental')
mus, i_sigmas, phi_sigmas = get_distr_params(encoder)
```

```python
def sinusoid(x):
    return 1/4 + 1/4*np.cos(x - pi)
```

```python
c = cm.rainbow(np.linspace(0, 1, M))



for neuron_type in range(M):
    n, bins, pathces = plt.hist(exp_phis[neuron_type], bins=20, density=True, color=0.7*c[neuron_type], alpha=0.3)

x_values = np.arange(0, 2*pi, 0.01)

plt.plot(x_values, sinusoid(x_values))
for neuron_type in range(M):
    phi_mean_exp = np.mean(exp_phis[neuron_type])
    phi_std_exp = np.std(exp_phis[neuron_type])
    phi_mean_th = mus[neuron_type]
    phi_std_th = phi_sigmas[neuron_type]
    
    print(phi_mean_th - phi_mean_exp)
    
    y_values_th = scipy.stats.norm(phi_mean_th, phi_std_th)
    y_values_exp = scipy.stats.norm(phi_mean_exp, phi_std_exp)
    
    plt.plot(x_values, y_values_th.pdf(x_values), color=c[neuron_type],linestyle='dashed', linewidth=2)
    plt.plot(x_values, y_values_exp.pdf(x_values), color=c[neuron_type], linewidth=2)

M = encoder.M
N = encoder.N
I_s = encoder.input_params["I_s"]
plt_fig_3b(M, N, I_s, c)

plt.savefig('Figures/Figure_3b.svg', dpi=600)
plt.savefig('Figures/Figure_3b.png', dpi=600)
```

```python
def exp_correction(f):
    tau_m = 24*ms
    return 1/np.tanh(1/(f*tau_m))
```

```python
noise_lims = [0.05, 0.4]
noise_fracs = np.linspace(noise_lims[0], noise_lims[1], 100)

input_params = {}
input_params["automatic_range"] = True
input_params["I_min"] = 75*pA
input_params["I_max"] = 130*pA

simulation_params["num_oscillations"] = 1
```

```python
all_mus = []
all_sigmas = []

input_params["I_min"], input_params["I_max"] = encoder.input_params["I_min_lock"], encoder.input_params["I_max_lock"]
input_params["automatic_range"] = False

for noise_frac in noise_fracs:
    model_params["noise_frac"] = noise_frac
    encoder = PhaseEncoder(num_ensembles=M, ensemble_size=N, model_params=model_params, 
                       oscillation_params=oscillation_params, input_params=input_params,
                       simulation_params=simulation_params, rnd_seed=0)

    mus, i_sigmas, phi_sigmas = get_distr_params(encoder)
    
    all_mus.append(mus[1:-1])
    all_sigmas.append(phi_sigmas[1:-1])

        
stoch_sigmas = list(map(lambda eta: model_sigma(eta, encoder), noise_fracs))
sigmas_matrix = np.array(all_sigmas).reshape((100, 98)).T
```

```python
noise_lims = [0.05, 0.4]
f_lims = [5*Hz, 15*Hz]
noise_fracs = np.linspace(noise_lims[0], noise_lims[1], 100)
fs = np.linspace(f_lims[0], f_lims[1], 3)
```

```python
M = 10
N = 1
simulation_params["num_oscillations"] = 3

input_params["automatic_range"] = True

experiments = itertools.product(noise_fracs, fs)
num_experiments = len(list(experiments))
experiments = itertools.product(noise_fracs, fs)

distribution_params = []
th_phis = []
all_exp_phis = []


for experiment, (noise_frac, f) in enumerate(experiments):
    model_params["noise_frac"] = noise_frac
    oscillation_params["f"] = f
    encoder = PhaseEncoder(num_ensembles=M, ensemble_size=N, model_params=model_params, 
                       oscillation_params=oscillation_params, input_params=input_params,
                       simulation_params=simulation_params, rnd_seed=0)
    
    if experiment%100==0:
        print("{} experiments completed out of {}".format(experiment, num_experiments))
    
    mus, i_sigmas, phi_sigmas = get_distr_params(encoder) 
    distribution_params.append((mus, phi_sigmas))
    #phis = run_simulation(encoder, mode='theoretical')
    #th_phis.append(phis)
    #exp_phis = run_simulation(encoder, mode='experimental_parallel')
    #all_exp_phis.append(exp_phis)

print("{} experiments completed out of {}".format(experiment+1, num_experiments))

        
stoch_sigmas = list(map(lambda eta: model_sigma(eta, encoder), noise_fracs))
```

```python
mutual_I = list(map(lambda params: approx_mutual_I(params, method='approx_second'), distribution_params))
#mutual_I = list(map(lambda phis: bins_mutual_I(phis), th_phis))
#mutual_I = list(map(lambda phis: bins_mutual_I(phis), all_exp_phis))
mI_matrix = np.array(mutual_I).reshape(100, 3)
```

```python
for i in [0, 1, 2]:
    freq_label = str(round(fs[i]/Hz))
    plt.plot(stoch_sigmas, mI_matrix[:, i], label=freq_label, linewidth=3)

plt_fig_3e(stoch_sigmas)
plt.savefig('Figures/Figure_3e.svg', dpi=600)
plt.savefig('Figures/Figure_3e.png', dpi=600)
```
