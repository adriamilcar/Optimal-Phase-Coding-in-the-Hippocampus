```python
M = 10
N = 100

model_params = {}
model_params["tau_m"] = 24*ms
model_params["R_m"] = 142e6*ohm
model_params["v_thres"] = 15*mV
model_params["v_rest"] = 0*mV
model_params["v_reset"] = 0*mV
model_params["tau_ref"] = 0*ms
model_params["v_0"] = 0*mV
model_params["noise_frac"] = 0.3
model_params["I_AH0"] = 0*pA
#model_params["noise_frac"] = 0.4


oscillation_params = {}
oscillation_params["I_osc"] = 40*pA
oscillation_params["f"] = 30*Hz

input_params = {}
input_params["I_min"] = 75*pA
input_params["I_max"] = 130*pA
input_params["automatic_range"] = True

simulation_params = {}
simulation_params["method"] = "euler"
simulation_params["num_oscillations"] = 10
simulation_params["monitor_spikes"] = True
simulation_params["monitor_voltage"] = False
simulation_params["dt"] = 0.05*ms
simulation_params["record_dt"] = 0.5*ms


encoder = PhaseEncoder(num_ensembles=M, ensemble_size=N, model_params=model_params, 
                       oscillation_params=oscillation_params, input_params=input_params,
                       simulation_params=simulation_params, rnd_seed=0)

run_simulation(encoder, mode='None')
```

```python
network_model = encoder

runtime = network_model.simulation_params["runtime"]
f = network_model.oscillation_params["f"]
spike_monitors = network_model.monitors["SM"]
first_spikes = network_model.groups['encoder'].lastspike + network_model.oscillation_params["oscillation_T"]
ftpds = first_passage_time(spike_monitors, f,
                            runtime, network_model.M, network_model.N, first_spikes)
```

```python
def first_passage_time(spike_monitor, freq, total_time, M, N, first_spikes):
    bin_time = 1/freq
    num_bins = int(total_time/bin_time)
    fptds = []
    for group in range(M):
        aux = []
        for neuron_index in range(group*N, group*N + N):
            neuron_times = list(spike_monitor.spike_trains().values())[neuron_index]
            try:
                index = np.where(neuron_times >= first_spikes[neuron_times])[0][0]
                fptd = neuron_times[index]
                aux.append(fptd)
            except: 
                pass
        fptds.append(aux)
    return fptds
```

```python
def get_passages(spike_monitors, M, N, min_time, passage_number):
    passage_times = np.zeros((M, N))
    key_list = ["SM_" + str(neuron) for neuron in range(M)]
    for input_, key in enumerate(key_list):
        skipped_neurons = []
        for neuron, spike_train in enumerate(spike_monitors[key].spike_trains().values()):
            try:
                index = np.where(spike_train >= min_time)[0][passage_number]
                passage_times[input_, neuron] = spike_train[index]
            except:
                skipped_neurons.append(neuron)
        passage_times[input_, skipped_neurons] = np.mean(passage_times[input_,:])
    return passage_times
```

```python
def v_of_t(v0, t, t_prime):
    A = 1/np.sqrt(1 + (tau*omega)**2)
    ph = -np.arctan(tau*omega) - pi
    delta_t = t - t_prime
    V0exp = v0*np.exp(-delta_t/tau)
    V_t = V0exp + R*I*(1 - np.exp(-delta_t/tau)) + R*Iosc*A*(np.cos(t*omega + ph) - np.exp(-delta_t/tau)*np.cos(t_prime*omega + ph))  
    return V_t

def var_of_t(t, t0):
    delta_t = t - t0
    var_w = (eta*(v_th - v_rest))**2
    exp_correction = (1 - np.exp(-2*delta_t/tau))
    var_v = var_w/2*exp_correction
    return var_v

def Gauss(x, x0, var):
    norm = 1/np.sqrt(2*pi*var)
    num = (x - x0)**2
    denom = 2*var
    return norm*np.exp(-num/denom)*dv
    
def p_v(v, t, v0, t0):
    mean = v_of_t(v0, t, t0)
    var = var_of_t(t, t0)
    prob = Gauss(v, mean, var)
    return prob
```

```python
tau = 24*ms
R = 142e6*ohm
f = 20*Hz
omega = 2*pi*f
Iosc = 40*pA
I = 110*pA
eta = 0.1
v_th = 15*mV
v_rest = 0*mV
dv = 1*mV

t_0 = 0*ms
max_t = 300*ms
#h is the timestep
#t_values[j] = j*h
h = 0.5*ms
t_values = np.arange(t_0, max_t, h)*second

sub_partition = 1
sub_h = h/sub_partition
st_values = np.arange(t_0, max_t, sub_h)*second

num_bins = len(t_values)
```

```python
def K(m, j):
    return p_v(v_th, t_values[m], v_th, t_values[j])   

def g(m):
    return p_v(v_th, t_values[m], 0*mV, 0*ms)

def r(m):
    sub_m = m*sub_partition
    sum_0 = 1/2*(p_v(v_th, st_values[sub_m], v_th, 0*ms) + p_v(v_th, st_values[sub_m], v_th, st_values[sub_m - 1]))
    int_sum = sum([p_v(v_th, st_values[sub_m], v_th, st_values[i]) for i in range(1, sub_m - 1)])
    return sub_h*(sum_0 + int_sum)


#def r(m):
#    v1 = v_of_t(t_values[m])
#    v2 = v_of_t(t_values[j])
#    return math.erfc(v2 - mean)/(2*sigma)) - math.erfc(v1 - mean)/(2*sigma))
    
K_matrix = np.zeros((num_bins, num_bins))
g_values = np.zeros(num_bins)
r_values = np.zeros(num_bins)*second
rho_values = np.zeros(num_bins)*Hz
denom_values = np.zeros(num_bins)
num_values = np.zeros(num_bins)

for timestep in range(1, num_bins):
    g_values[timestep] = g(timestep)
    r_values[timestep] = r(timestep)
    for timestep_prime in range(timestep):
        K_matrix[timestep][timestep_prime] = K(timestep, timestep_prime)
```

```python
for m in range(1, num_bins):
    k_rho_int = sum([K_matrix[m,j]*h*rho_values[j] for j in range(1, m)])
    k_int = sum([K_matrix[m,j] for j in range(1, m)])
        
    num = k_rho_int - g_values[m]
    denom = 1/2*h*K_matrix[m, 0] + h*k_int - r_values[m]

    rho_values[m] = num/denom
#    denom_values[m] = denom
#    num_values[m] = num
```
